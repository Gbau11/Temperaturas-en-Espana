```python
import os
import pandas as pd

carpeta = '../data/SP'

nombres_columna = ['estacion', 'fecha', 'maxima', 'minima', 'precip', 'nevada', 'profun_nieve']

dfs = []

for archivo in os.listdir(carpeta):
    if archivo.endswith('.csv'):
        ruta_completa = os.path.join(carpeta, archivo)
        
        df = pd.read_csv(ruta_completa, header=None, sep=';')
        
        df.columns = nombres_columna
        
        dfs.append(df)
```

```python
df = pd.concat(dfs, ignore_index=True)
df
```

```python
df['precip'] = df['precip'].apply(lambda x: x / 10 if pd.notnull(x) else x)
df
```

```python
df['fecha'] = pd.to_datetime(df['fecha'])
```

```python
df['año'] = df['fecha'].dt.year
df['mes'] = df['fecha'].dt.month
```

```python
gb = df.groupby(['estacion', df['año'], df['mes']]).agg({'precip': 'sum', 'profun_nieve': 'sum'})
gb
```
